```python
import re
postal_code_regex = re.compile(r"\bH\d\w\b")
zipcode_regex = re.compile(r"\d{5} ")

region_regex = re.compile(r"[A-Z]{2}")

all_regex = re.compile(r"\bH\d\w\b | \d{5}")
```

```python
def find_postal_code(s):
    matches = postal_code_regex.findall(s)
    return matches[0] if matches else None

def find_zipcode(s):
    matches = zipcode_regex.findall(s)
    return matches[0] if matches else None

def find_region(s):
    matches = region_regex.findall(s)
    return matches[0] if matches else None

def find_all(s):
    matches = all_regex.findall(s)
    return matches[0] if matches else None
```

```python
b['region'] = b.full_address.map(find_region)
```
