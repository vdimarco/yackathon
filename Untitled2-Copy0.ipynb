```python
def process_activity(df, year):
    by_codes = df[df.year == year].groupby('codes')
    year = str(year)
    codes = df.codes.unique()
    activity = pd.DataFrame(index=codes)
    activity['business_activity_' + year] = by_codes['business_id'].nunique()
    activity['user_activity_' + year] = by_codes['user_id'].nunique()
    activity['review_activity_' + year] = by_codes.count()['codes']
    activity['mean_stars_' + year] = by_codes['stars'].mean()
    activity['std_stars_' + year] = by_codes['stars'].std()
    return activity

                                           
```
